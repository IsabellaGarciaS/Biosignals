En la multiplicación de dos vectores, como se puede observar en la imagen de abajo, se multiplica los elemenos de las mismas posiciones y se van sumando.

image.png

j) Construyala función  𝑦[𝑛]=sin(𝜋∗0.18𝑛)en el intervalo0 <=n <=80.

```python
import numpy as np
import matplotlib.pyplot as plt

# Definir la función y[n]
def calculate_y(n):
    return np.sin(np.pi * 0.18 * n)

# Crear un arreglo de valores n en el intervalo 0 <= n <= 80
n_values = np.arange(0, 81)

# Calcular y[n] utilizando la función definida y un bucle
y_values = [calculate_y(n) for n in n_values]

# Graficar la función
plt.plot(n_values, y_values, label=r'$y[n] = \sin(\pi \times 0.18 \times n)$')
plt.xlabel('n')
plt.ylabel('y[n]')
plt.title('Gráfica de la función y[n]')
plt.legend()
plt.grid(True)
plt.show()
```

k) Construya otra senoidal𝑦2[𝑛]=cos(2𝜋∗0.03𝑛)

```python
#Se va a asumir que 0<=n<=80

# Definir la función y2[n]
def generate_y2(n_values):
    return np.cos(2 * np.pi * 0.03 * n_values)

# Crear un arreglo de valores n en el intervalo 0 <= n <= 80
n_values = np.arange(0, 81)

# Calcular y2[n] utilizando la función definida
y2_values = generate_y2(n_values)

# Graficar la función
plt.plot(n_values, y2_values, label=r'$y_2[n] = \cos(2\pi \times 0.03 \times n)$')
plt.xlabel('n')
plt.ylabel('y2[n]')
plt.title('Gráfica de la función y2[n]')
plt.legend()
plt.grid(True)
plt.show()
```

l) Genere una    tercera    señal    que    sea    la    suma    de    estas    dos    señales,    es    decir 𝑠[𝑛]=y[n]+𝑦2[𝑛]y una cuarta, que sea el producto de estas, 𝑡[𝑛]=y[n].𝑦2[𝑛].

```python
# Definir el rango de muestras
n = np.arange(81)

# Definir la primera señal y[n] = sin(pi*0.18*n)
y = np.sin(np.pi * 0.18 * n)

# Definir la segunda señal y2[n] = cos(2*pi*0.03*n)
y2 = np.cos(2 * np.pi * 0.03 * n)

# Generar la tercera señal s[n] = y[n] + y2[n]
s = y + y2

# Generar la cuarta señal t[n] = y[n] * y2[n]
t = y * y2

# Graficar las señales
plt.figure(figsize=(12, 8))

# Gráfico de y[n]
plt.subplot(2, 2, 1)
plt.plot(n, y)
plt.title('Señal y[n] = sin(pi*0.18*n)')
plt.xlabel('n')
plt.ylabel('Amplitud')

# Gráfico de y2[n]
plt.subplot(2, 2, 2)
plt.plot(n, y2)
plt.title('Señal y2[n] = cos(2*pi*0.03*n)')
plt.xlabel('n')
plt.ylabel('Amplitud')

# Gráfico de s[n]
plt.subplot(2, 2, 3)
plt.plot(n, s)
plt.title('Señal s[n] = y[n] + y2[n]')
plt.xlabel('n')
plt.ylabel('Amplitud')

# Gráfico de t[n]
plt.subplot(2, 2, 4)
plt.plot(n, t)
plt.title('Señal t[n] = y[n] * y2[n]')
plt.xlabel('n')
plt.ylabel('Amplitud')

plt.tight_layout()
plt.show()
```

Escribir una función que reciba un diccionario con las notas de los alumnos de un curso y devuelva una serie con la nota mínima, la máxima, media y la desviación típica.

```python
import pandas as pd
import numpy as np

#Se crea una función para ayudar a determinar las notas máximas, mínimas, la media y la desviasión típica
def calcular_estadisticas(notas_alumnos):
    # Crear una serie a partir del diccionario de notas
    notas_serie = pd.Series(notas_alumnos)

    # Calcular estadísticas
    nota_minima = notas_serie.min()
    nota_maxima = notas_serie.max()
    nota_media = notas_serie.mean()
    desviacion_tipica = notas_serie.std()

    # Crear una serie con las estadísticas calculadas
    estadisticas = pd.Series({
        'Nota mínima': nota_minima,
        'Nota máxima': nota_maxima,
        'Media': nota_media,
        'Desviación típica': desviacion_tipica
    })

    return estadisticas

# Verificando que la función esté buena se colocaron los siguientes alumnos con notas inventadas
notas_curso = {
    'JUAN': 4.5,
    'CARLOS': 4.9,
    'MARIA': 3.8,
    'JUANA': 2.5,
    'MARIANA': 4.8,
    'MARIA JOSÉ': 3.3
}

resultados = calcular_estadisticas(notas_curso)
print(resultados)
```
