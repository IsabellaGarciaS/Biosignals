j) Construyala funci√≥n  ùë¶[ùëõ]=sin(ùúã‚àó0.18ùëõ)en el intervalo0 <=n <=80.

```python
import numpy as np
import matplotlib.pyplot as plt

# Definir la funci√≥n y[n]
def calculate_y(n):
    return np.sin(np.pi * 0.18 * n)

# Crear un arreglo de valores n en el intervalo 0 <= n <= 80
n_values = np.arange(0, 81)

# Calcular y[n] utilizando la funci√≥n definida y un bucle
y_values = [calculate_y(n) for n in n_values]

# Graficar la funci√≥n
plt.plot(n_values, y_values, label=r'$y[n] = \sin(\pi \times 0.18 \times n)$')
plt.xlabel('n')
plt.ylabel('y[n]')
plt.title('Gr√°fica de la funci√≥n y[n]')
plt.legend()
plt.grid(True)
plt.show()
```

k) Construya otra senoidalùë¶2[ùëõ]=cos(2ùúã‚àó0.03ùëõ)

```python
#Se va a asumir que 0<=n<=80

# Definir la funci√≥n y2[n]
def generate_y2(n_values):
    return np.cos(2 * np.pi * 0.03 * n_values)

# Crear un arreglo de valores n en el intervalo 0 <= n <= 80
n_values = np.arange(0, 81)

# Calcular y2[n] utilizando la funci√≥n definida
y2_values = generate_y2(n_values)

# Graficar la funci√≥n
plt.plot(n_values, y2_values, label=r'$y_2[n] = \cos(2\pi \times 0.03 \times n)$')
plt.xlabel('n')
plt.ylabel('y2[n]')
plt.title('Gr√°fica de la funci√≥n y2[n]')
plt.legend()
plt.grid(True)
plt.show()
```

l) Genere una    tercera    se√±al    que    sea    la    suma    de    estas    dos    se√±ales,    es    decir ùë†[ùëõ]=y[n]+ùë¶2[ùëõ]y una cuarta, que sea el producto de estas, ùë°[ùëõ]=y[n].ùë¶2[ùëõ].

```python
# Definir el rango de muestras
n = np.arange(81)

# Definir la primera se√±al y[n] = sin(pi*0.18*n)
y = np.sin(np.pi * 0.18 * n)

# Definir la segunda se√±al y2[n] = cos(2*pi*0.03*n)
y2 = np.cos(2 * np.pi * 0.03 * n)

# Generar la tercera se√±al s[n] = y[n] + y2[n]
s = y + y2

# Generar la cuarta se√±al t[n] = y[n] * y2[n]
t = y * y2

# Graficar las se√±ales
plt.figure(figsize=(12, 8))

# Gr√°fico de y[n]
plt.subplot(2, 2, 1)
plt.plot(n, y)
plt.title('Se√±al y[n] = sin(pi*0.18*n)')
plt.xlabel('n')
plt.ylabel('Amplitud')

# Gr√°fico de y2[n]
plt.subplot(2, 2, 2)
plt.plot(n, y2)
plt.title('Se√±al y2[n] = cos(2*pi*0.03*n)')
plt.xlabel('n')
plt.ylabel('Amplitud')

# Gr√°fico de s[n]
plt.subplot(2, 2, 3)
plt.plot(n, s)
plt.title('Se√±al s[n] = y[n] + y2[n]')
plt.xlabel('n')
plt.ylabel('Amplitud')

# Gr√°fico de t[n]
plt.subplot(2, 2, 4)
plt.plot(n, t)
plt.title('Se√±al t[n] = y[n] * y2[n]')
plt.xlabel('n')
plt.ylabel('Amplitud')

plt.tight_layout()
plt.show()
```

Escribir una funci√≥n que reciba un diccionario con las notas de los alumnos de un curso y devuelva una serie con la nota m√≠nima, la m√°xima, media y la desviaci√≥n t√≠pica.

```python
import pandas as pd
import numpy as np

#Se crea una funci√≥n para ayudar a determinar las notas m√°ximas, m√≠nimas, la media y la desviasi√≥n t√≠pica
def calcular_estadisticas(notas_alumnos):
    # Crear una serie a partir del diccionario de notas
    notas_serie = pd.Series(notas_alumnos)

    # Calcular estad√≠sticas
    nota_minima = notas_serie.min()
    nota_maxima = notas_serie.max()
    nota_media = notas_serie.mean()
    desviacion_tipica = notas_serie.std()

    # Crear una serie con las estad√≠sticas calculadas
    estadisticas = pd.Series({
        'Nota m√≠nima': nota_minima,
        'Nota m√°xima': nota_maxima,
        'Media': nota_media,
        'Desviaci√≥n t√≠pica': desviacion_tipica
    })

    return estadisticas

# Verificando que la funci√≥n est√© buena se colocaron los siguientes alumnos con notas inventadas
notas_curso = {
    'JUAN': 4.5,
    'CARLOS': 4.9,
    'MARIA': 3.8,
    'JUANA': 2.5,
    'MARIANA': 4.8,
    'MARIA JOS√â': 3.3
}

resultados = calcular_estadisticas(notas_curso)
print(resultados)
```
